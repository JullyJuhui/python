```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rc('axes', unicode_minus=False)
```

```python
import pandas as pd
df = pd.read_excel('출생아수.xlsx', skiprows=2, nrows=2, index_col=0)  #첫번째 col: 0
df
```

```python
df.index.values
```

```python
df.rename(index={'출생아\xa0수': '출생아수'}, inplace=True)
df.rename(index={'합계\xa0출산율': '합계출산율'}, inplace=True)
```

```python
df = df.T
```

```python
df.index.name = '년도'
df.reset_index(inplace=True)
df
```

```python
# plt.bar(df.index, df['출생아수'], color='skyblue', hatch='..', ec='w')
plt.plot(df.index, df['출생아수'])
plt.plot(df.index, df['합계출산율'])

plt.show()
```

```python
fig, ax1 = plt.subplots(figsize = (10, 5))

ax1.bar(df['년도'], df['출생아수'], color='skyblue', hatch='..', ec='w')
ax1.set_ylabel('출생아수', color='green')
ax1.set_ylim(150, 550)
for idx, y in enumerate(df['출생아수']):
    ax1.text(idx, y+2, f'{y:,}', ha='center', color='brown')

ax2 = ax1.twinx()
ax2. plot(df.index, df['합계출산율'], color='deeppink', marker='*', ms=25, lw=5, mec='w', mew = 3)
ax2.set_ylabel('합계출산율', color='green')
ax2.set_ylim(0, 1.5)
for idx, y in enumerate(df['합계출산율']):
    ax2.text(idx, y+0.05, f'{y:.2f}', ha='center', color='brown')

plt.show()
```
