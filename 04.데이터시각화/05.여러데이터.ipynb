```python
import matplotlib.pyplot as plt

#[한글 설정]
plt.rc('font', family = 'Malgun Gothic')  #글꼴
plt.rc('font', size = 10)  #크기

plt.rc('axes', unicode_minus = False)  #마이너스
```

```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')
```

```python
kor = list(df['국어'])
eng = list(df['영어'])
mat = list(df['수학'])
name = list(df['이름'])
```

```python
plt.figure(figsize=(10, 4))

plt. plot(name, kor, label = '국어', color = 'red', marker = 'o', ls='--')
plt.plot(name, eng, label = '영어', color = 'blue', marker = 's', ls = '-.')
plt.plot(name, mat, label = '수학', color = 'green', marker = 'X')

plt.legend(loc=(1.02, 0.5))

for idx, k in enumerate(kor):  #국어
    plt.text(name[idx], kor[idx]+2, k, color = 'red', ha='center', size=10)

for idx, k in enumerate(eng):  #영어
    plt.text(name[idx], eng[idx]+2, k, color = 'blue', ha='center', size=10)

for idx, k in enumerate(mat):  #수학
    plt.text(name[idx], mat[idx]+2, k, color = 'green', ha='center', size=10)

plt.show()
```
