```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.rc('axes', unicode_minus=False)

import pandas as pd
df = pd.read_csv('data/score.csv')
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
plt.scatter(df['국어'], df['영어'], sizes = df['학년']*500, c = df['학년'],
            cmap='spring', alpha=0.5)
plt.xlabel('국어')
plt.ylabel('영어')

plt.colorbar(ticks=[1, 2, 3], label='학년', orientation = 'horizontal', shrink = 1.1)  #orientation = 'horizontal': 가로

plt.show()
```

```python
plt.scatter(df['국어'], df['영어'], sizes = df['학년']*500, c = df['학년'],
            cmap='spring', alpha=0.5)
plt.xlabel('국어')
plt.ylabel('영어')

plt.colorbar(ticks=[1, 2, 3], label='학년', orientation = 'horizontal', shrink = 1.1)

for idx in range(len(df)):
    x = df.loc[idx, '국어']
    y = df.loc[idx, '영어']
    name = df.loc[idx, '이름']
    plt.text(x, y-2, name, ha='center', color = 'brown')

plt.show()
```

```python
plt.scatter(df['수학'], df['과학'], sizes = df['학년']*500, c = df['학년'],
            cmap='spring', alpha=0.5)
plt.xlabel('수학점수')
plt.ylabel('과학점수')

plt.colorbar(ticks=[1, 2, 3], label='학년', orientation = 'horizontal', shrink = 0.6)

for idx in range(len(df)):
    x = df.loc[idx, '수학']
    y = df.loc[idx, '과학']
    name = df.loc[idx, '이름']
    plt.text(x, y-1, name, ha='center', color = 'brown')

plt.show()
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
plt.scatter(df['국어'], df['영어'], s=df['학년']*500, c = df['학년'], cmap='spring', alpha=0.5)
plt.colorbar(ticks=[1, 2, 3], orientation='horizontal', label='학년', shrink=0.7)

for idx in range(len(df)):
    x = df.loc[idx, '국어']
    y = df.loc[idx, '영어']
    name = df.loc[idx, '이름']
    plt.text(x, y-1.5, name, ha='center', color = 'brown')

plt.ylim(0, 120)

plt.show()
```
