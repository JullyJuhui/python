```python
import matplotlib.pyplot as plt
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.rc('axes', unicode_minus=False)
```

```python
import pandas as pd
df = pd.read_csv('data/score.csv')
```

학생별(이름) 국어, 영어, 수학점수 - 꺾은선 그래프

```python
plt.figure(figsize=(10, 3))
plt.title('학생별 국어, 영어점수', size = 17)
plt.xlabel('이름', size=13, loc='right', color='deeppink')
plt.ylabel('점수', size=13, loc='top', color='deeppink')

plt.plot(df['이름'], df['국어'], label='국어',
         marker='o', ms = 8, mec = 'hotpink', mew=3, mfc = 'pink')
plt.plot(df['이름'], df['영어'], label='영어',
         lw = 3, ls = '--', color = 'green')

plt.legend(loc=(0.38,-0.28), ncol=2)  #범례 길게 - ncol=2

yticks = [y for y in range(0, 101, 50)]
plt.yticks(yticks, [f"{y}점" for y in yticks])
plt.ylim(0, 110)
plt.xticks(df['이름'], [x+'님' for x in df['이름']])

plt.show()
```

학생별(이름) 키 막대그래프

```python
plt.figure(figsize=(10, 3.5))
plt.bar(df['이름'], df['키'])

plt.bar(df['이름'], df['키'], color='white', hatch='*', ec = 'orange')

yticks = [y for y in range(160, 211, 10)]
plt.yticks(yticks, [f'{y}cm' for y in yticks])
plt.ylim(160, 210)

xticks = [f"{df.loc[idx, '이름']}\n{df.loc[idx, '지원번호']}" for idx in range(len(df))]
plt.xticks(df['이름'], xticks)

plt.show()
```

학생별 평균점수를 막대그래프

```python
df['평균'] = df.apply(lambda row: row['국어':'사회'].mean(), axis=1)

plt.bar(df['이름'], df['평균'])

for idx, avg in enumerate(df['평균']):
    plt.text(idx, avg+1, f'{avg:.2f}', ha='center')

plt.ylim(0, 100)

plt.show()
```

```python
group = df.groupby('SW특기').size()
plt.bar(group.index, group.values)
plt.show()
```

누적 막대그래프

```python
index = list(df.index)  # index = [idx for idx in range(len(df))]
df.loc[0, '국어'] + df.loc[0, '영어'] + df.loc[0, '수학']
```

```python
plt.bar(df['이름'], df['국어'], label='국어')
plt.bar(df['이름'], df['영어'], bottom=df['국어'], label='영어')
plt.bar(df['이름'], df['수학'], bottom=df['국어']+df['영어'], label='수학')

plt.ylim(0, 300)
plt.legend()

for idx in df.index:
    sum = df.loc[idx, '국어'] + df.loc[idx, '영어'] + df.loc[idx, '수학']
    plt.text(idx, sum+3, f"{sum}점", ha='center')

plt.show()
```

```python
#학생별 사회, 과학 - 누적막대그래프
plt.bar(df['이름'], df['사회'], label = '사회')
plt.bar(df['이름'], df['과학'], bottom = df['사회'], label = '과학')

plt.legend(loc=(1.02, 0.5))

plt.ylim(0, 200)

index = list(df.index)
for idx in index:
    sum = df.loc[idx, '사회'] + df.loc[idx, '과학']
    plt.text(idx, sum+2, f"{sum}점", ha='center')

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.rc('axes', unicode_minus=False)

import pandas as pd
df = pd.read_csv('data/score.csv')
```

```python
plt.bar(df['이름'], df['국어'], label='국어', hatch='//', ec='white')
plt.bar(df['이름'], df['영어'], bottom=df['국어'], label='영어', hatch='.', ec='white')
plt.bar(df['이름'], df['수학'], bottom=df['국어']+df['영어'], label='수학', hatch='xx', ec='white')

plt.ylim(0, 300)
plt.legend()

plt.show()
```
