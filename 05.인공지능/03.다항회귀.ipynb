```python
def model_poly(degree):
    dataset = pd.read_csv('data/PolynomialRegressionData.csv')
    dataset.head()

    X = dataset.iloc[:, :-1].values
    y = dataset.iloc[:, -1].values

    from sklearn.preprocessing import PolynomialFeatures
    poly = PolynomialFeatures(degree=degree)  #n
    X_poly = poly.fit_transform(X)  #, x, x^2, ...x^n

    poly_reg = LinearRegression()
    poly_reg.fit(X_poly, y)

    return poly_reg
```

```python
model = model_poly(2)
```

```python
#4시간 공부 점수
x = poly.fit_transform([[5]])
model.predict(x)
```

```python
def graph(degree):
    model = model_poly(degree)

    import numpy as np
    X_range = np.arange(np.min(X), np.max(X), 0.1)
    X_range = X_range.reshape(len(X_range), 1)

    from sklearn.preprocessing import PolynomialFeatures
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X_range)

    plt.figure(figsize=(5, 3))
    plt.scatter(X, y, label='실제점수')
    plt.plot(X_range, model.predict(X_poly), c='orange', label='예측점수')

    plt.legend()
    plt.show()
```

```python
graph(10)
```

```python
def predict(hour):
    model = model_poly(2)
    x = poly.fit_transform([[hour]])
    pred = model.predict(x)
    print(f"다항회귀: {hour}시간 공부했을 경우 예상점수는 '{pred[0]:.2f}점'입니다.")

    reg = LinearRegression()
    reg.fit(X, y)
    pred = reg.predict([[hour]])
    print(f"선형회귀: {hour}시간 공부했을 경우 예상점수는 '{pred[0]:.2f}점'입니다.")
```
