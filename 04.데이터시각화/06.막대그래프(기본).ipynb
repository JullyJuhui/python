```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')

import matplotlib.pyplot as plt

#[한글 설정]
plt.rc('font', family = 'Malgun Gothic')  #글꼴
plt.rc('font', size = 10)  #크기

plt.rc('axes', unicode_minus = False)  #마이너스
```

```python
name = df['이름']
height = df['키']
```

```python
colors = ['r', 'orange', 'yellow', 'g', 'skyblue', 'c', 'b', 'pink']

plt.bar(name, height, color = colors, alpha=0.2)
plt.show()
```

```python
plt.bar(name, height, width=0.5)

plt.yticks([y for y in range(160, 230, 5)])
plt.ylim(160, 210)

#x축 눈금 세로로 출력
plt.xticks(name, df.index, rotation = 90)  #name자리에 index를 줘라~~/ 90도

plt.show()
```
