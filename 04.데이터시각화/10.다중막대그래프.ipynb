```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.rc('axes', unicode_minus=False)

import pandas as pd
df = pd.read_csv('data/score.csv')
```

학생별 국어, 영어 막대그래프

```python
# 한 사람당 4칸씩 차지해야함 - 국, 영, 수, 빈칸
# (n-1)/2
x = df.index * 4
plt.bar(x + 0, df['국어'], label = '국어', width=1, color='hotpink', hatch='.', ec='white')
plt.bar(x + 1, df['영어'], label = '영어', width=1, color='green', hatch='.', ec='white')
plt.bar(x + 2, df['수학'], label = '수학', width=1, color='skyblue', hatch='.', ec='white')

xticks = [idx+1 for idx in x]
plt.xticks(xticks, df['이름'])
plt.legend(ncol = 3, loc=(0.2, -0.15))

for idx, score in enumerate(df['국어']):
    plt.text(idx*4+0, score+0.5, score, ha='center', size=9)

for idx, score in enumerate(df['영어']):
    plt.text(idx*4+1, score+0.5, score, ha='center', size=9)

for idx, score in enumerate(df['수학']):
    plt.text(idx*4+2, score+0.5, score, ha='center', size=9)

plt.show()
```

```python
#사회, 과학

x = df.index * 3
plt.bar(x+0, df['사회'], label = '사회', width=1, color='hotpink', hatch='.', ec='white')
plt.bar(x+1, df['과학'], label = '과학', width=1, color='skyblue', hatch='.', ec='white')

plt.legend()

for idx, score in enumerate(df['사회']):
    plt.text(idx*3+0, score, score, ha='center')

for idx, score in enumerate(df['과학']):
    plt.text(idx*3+1, score, score, ha='center')

xticks = [idx+0.5 for idx in x]
plt.xticks(xticks, df['이름']+'\n' + df['지원번호'])

plt.show()
```

```python
x = df.index * 4
plt.figure(figsize=(10, 5))
plt.bar(x, df['국어'], width=1, label='국어')
plt.bar(x+1, df['영어'], width=1, label='영어')
plt.bar(x+2, df['수학'], width=1, label='수학')

plt.legend()

xticks = [t+1 for t in x]
plt.xticks(xticks, df['이름'])

plt.show()
```
