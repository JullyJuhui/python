```python
from konlpy.tag import Okt
from collections import Counter
```

```python
with open('애국가.txt', 'r', encoding='utf-8') as file:
    text = file.read()

print(text)
```

```python
okt = Okt()
nouns = okt.nouns(text)
nouns = [noun for noun in nouns if len(noun) > 1]  #한글자 단어 제거
```

```python
counter = Counter(nouns)
```

```python
import numpy as np
from PIL import Image

heart_mask = np.array(Image.open('heart_mask.png'))

with open('hardtolove.txt', 'r') as file:
    text = file.read()
```

```python
wc = WordCloud(width=800, 
               height=800,
               font_path='GowunDodum-Regular.ttf',
               scale=3.0,
               mask=heart_mask,
               background_color="#C2EAFA"
               ).generate_from_frequencies(counter)
```

```python
plt.imshow(wc)
plt.axis('off')

plt.savefig('애국가.png')
plt.show()
```
