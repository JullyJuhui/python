```python
import matplotlib.pyplot as plt
```

```python
#[한글 설정]
plt.rc('font', family = 'Malgun Gothic')  #글꼴
plt.rc('font', size = 10)  #크기

plt.rc('axes', unicode_minus = False)  #마이너스
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 6, 9, 12, 15]
```

```python
plt.plot(x, y)

#눈금
plt.xticks([x for x in range(min(x), max(x)+1, 1)], color = 'blue')
plt.yticks([y for y in range(min(y), max(y)+1, 2)], color = 'blue')
plt.show()
```
