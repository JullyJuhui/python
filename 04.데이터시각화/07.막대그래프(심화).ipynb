```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')

import matplotlib.pyplot as plt

#[한글 설정]
plt.rc('font', family = 'Malgun Gothic')  #글꼴
plt.rc('font', size = 10)  #크기

plt.rc('axes', unicode_minus = False)  #마이너스
```

```python
name = df['이름']
height = df['키']
```

```python
plt.barh(name, height)

plt.ylabel('학생이름')
plt.xlabel('키')

plt.xlim(160, 210)

plt.show()
```

```python
plt.figure(figsize=(10, 3))
#plt.bar(name, height, color = 'none', ec = 'green', hatch='.')  #/, x, ., .., o
bar = plt.bar(name, height, color = 'none', ec = 'green')
plt.ylim(160, 210)

bar[0].set_hatch('.')

for idx, h in enumerate(height):
    plt.text(idx, h+1, h, ha='center')

plt.show()
```

```python
#학생별 국어점수
plt.figure(figsize=(10, 5))

name = list(df['이름'])
avg = list((df['국어'] + df['영어'] + df['수학'] + df['사회'] + df['과학'])/5)
bar = plt.bar(name, avg, color = 'none', ec = 'brown', hatch = '/')
# bar[0].set_hatch('.')
# plt.bar(name, avg)

plt.yticks([y for y in range(0, 101, 5)])
plt.ylim(0, 100)

plt.title('학생별 평균', size = 15)
plt.xlabel('학생이름')
plt.ylabel('평균점수')

for idx, avg in enumerate(avg):
    plt.text(idx, avg+1, f'{avg:.02f}', ha='center')

plt.show()
```

```python
import numpy as np

def graph(name):
    filt = df['이름']==name
    df2 = df[filt]
    if len(df2)==0:
        return np.nan
    
    else:
        df3 = df2.loc[:, '국어':'사회']
        df4 = df3.iloc[0]
        plt.title(f'{name}의 성적표', size=17)
        plt.bar(df4.index, df4.values)
        plt.ylim(0, 100)
        return plt

img = graph('송태섭')

if pd.isna(img):
    print('없음')
else:
    plt.show()
```
