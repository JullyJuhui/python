```python
import matplotlib.pyplot as plt

#[한글 설정]
plt.rc('font', family = 'Malgun Gothic')  #글꼴
plt.rc('font', size = 10)  #크기

plt.rc('axes', unicode_minus = False)  #마이너스
```

```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')
df
```

```python
name = list(df['이름'])
height = list(df['키'])

name, height  #x, y
```

그래프 그리기

```python
#제목
plt.title('학생별 신장\n', size = 18)

#축 레이블
plt.xlabel('이름\n', color = 'blue', size = 13)
plt.ylabel('키(cm)\n', color = 'blue', size = 13)

#눈금
plt.yticks([y for y in range(160, 220, 5)])  


plt.plot(name, height, linewidth = '2', marker='s')  
#linewidth: 그래프 선의 굵기
#marker: X, x, v, o, >, s
plt.show()
```

```python
plt.title('학생별 신장\n', size = 18)

plt.xlabel('이름\n', color = 'blue', size = 13)
plt.ylabel('키(cm)\n', color = 'blue', size = 13)

plt.yticks([y for y in range(160, 220, 5)])  

plt.plot(name, height, 
         linewidth = '2', linestyle = ':', color = 'brown', #line  /style: -, --, -., :, none
         marker='s', markersize = 6, markerfacecolor='pink', #marker
         markeredgecolor = 'brown', markeredgewidth = 1)  #marker edge
plt.show()
```

```python
plt.title('학생별 신장\n', size = 18)

plt.xlabel('이름\n', color = 'blue', size = 13)
plt.ylabel('키(cm)\n', color = 'blue', size = 13)

plt.yticks([y for y in range(160, 220, 5)]) 

#축약 가능
plt.plot(name, height,
         lw = 5, c = 'brown', ls = '--', 
         marker = 'o', ms = 20, mfc = "#e46ebf", mec='green', mew=5)  

plt.show()
```

```python
plt.figure(figsize=(10, 3))  #그래프 크기
plt.ylim(160, 220) #y limit
plt.plot(name, height, marker='o')

#텍스트 추가
for idx,txt in enumerate(height):
    plt.text(name[idx], height[idx]+1, f'{txt}cm', size = 8, ha = 'center')

plt.show()
```
