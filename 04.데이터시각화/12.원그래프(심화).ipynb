```python
enroll = pd.read_csv('data/수강.csv')
enroll.head()
```

```python
group = enroll.groupby('강좌번호').size().to_frame('인원수')
cou = pd.read_csv('data/강좌.csv')
merge = cou.merge(group, how='inner', on='강좌번호')
```

```python
labels = merge['강좌이름']
values = merge['인원수']
```

```python
colors = ['#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF']
plt.pie(values, labels=labels, wedgeprops={'width':0.6, 'edgecolor':'black'},
        autopct='%.1f%%', explode=[0.05]*len(values), colors=colors)
plt.title('강좌 수강비율', size=17)

plt.show()
```

학점 비율

```python
def grade(score):
    if score >= 90:
        return 'A'
    
    elif score >= 80:
        return 'B'
    
    elif score >= 70:
        return 'C'
    
    elif score >= 60:
        return 'D'
    
    else:
        return'F'
```

```python
enroll = pd.read_csv('data/수강.csv')
enroll['학점'] = enroll['점수'].apply(grade)

group = enroll.groupby('학점').size()
```

```python
plt.pie(group.values, labels=group.index + '학점', 
        autopct='%.0f%%', explode=[0.05]*len(group),
        textprops={'color':'pink'},
        wedgeprops={'width':0.5}, counterclock=False, startangle=150)

plt.show()
```
