```python
import matplotlib.pyplot as plt

#[한글 설정]
plt.rc('font', family = 'Malgun Gothic')  #글꼴
plt.rc('font', size = 10)  #크기

plt.rc('axes', unicode_minus = False)  #마이너스
```

```python
year = [y for y in range (2020, 2025, 1)]  #연속적인 값 - for문 이용
height = [ 145, 150, 160, 165, 173]
```

```python
plt.plot(year, height, label = '연도별 신장')

#범례 설정: default = upper left
plt.legend(loc='lower right')  #(upper/lower) + right/left
plt.legend(loc=(1.05, 0.7))  #직접 숫자로 위치 지정 -> 범례를 바깥으로 보낼 수 있음

plt.show()
```
